```python
import requests
import pandas as pd
import time
import folium
```

```python
iss = "https://api.wheretheiss.at/v1/satellites/25544"
```

```python
positions = []
for i in range(20):
    response = requests.get(iss)
    data = response.json()
    positions.append(data)
    time.sleep(5)
```

```python
df = pd.DataFrame(positions)
```

```python
lat = list(df["latitude"])
```

```python
long = list(df["longitude"])
```

```python
pos = dict()
for i in range(len(lat)):
    pos[lat[i]]= long[i]
```

```python
world_map = folium.Map(location=[df["latitude"][0], df["longitude"][0]], zoom_start=1)
```

```python
for x, y in pos.items():
    folium.Marker([x, y]).add_to(world_map)
```
